```python
def gradient_descent( lr =0.1):
    max_itr = 50
    x = 0
    err_list = []
    x_values = []
    for i in range(max_itr):
        x_values.append(x)
        e = (x-5)**2
        err_list.append(e)
        # Update Rule
        x = x - lr*2*(x-5)
        
    return x_values, err_list, x
```

```python
x_values, err_list, x = gradient_descent()
```
