```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

```

```python
!find / -name dogs.jpeg
```

```python
# Open the image
img = Image.open('/content/dogs.jpeg')
```

```python
# Display the original image
plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')
```

```python
# Convert the image to a numpy array
img_array = np.array(img)
```

```python
# Get the dimensions of the image
w, h, d = img_array.shape
```

```python
# Reshape the image array to a 2D array of pixels
img_array_2d = img_array.reshape(w * h, d)
```

```python
# Apply K-Means clustering with k = 3
kmeans = KMeans(n_clusters=3)
kmeans.fit(img_array_2d)

# Predict cluster labels for each pixel
labels = kmeans.labels_









```

```python
import matplotlib.pyplot as plt

# Display the original image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

# Display the clustered image
plt.subplot(1, 2, 2)
plt.imshow(labels.reshape(img_array.shape[0], img_array.shape[1]), cmap='viridis')
plt.title('Clustered Image')
plt.axis('off')
plt.show()
```

```python
# Reshape the labels to match the original image shape
labels_reshaped = labels.reshape(w, h)
```

```python
# Plot the clustered image
plt.subplot(1, 2, 2)
plt.imshow(labels_reshaped)
plt.title('Clustered Image')
plt.axis('off')
plt.show()
```

```python
# Get the cluster centers (dominant colors)
dominant_colors = kmeans.cluster_centers_
```

```python
# Convert the dominant colors to uint8
dominant_colors = dominant_colors.astype('uint8')
```

```python
# Display the dominant colors
plt.figure(figsize=(8, 2))
for i, color in enumerate(dominant_colors):
    plt.subplot(1, 3, i + 1)
    plt.imshow(np.array([[color]]))
    plt.axis('off')
plt.show()
```
